```python
from cnn import cnn
import numpy as np
```

```python
input_size = 3
num_output = 2


num_kernels = 6
kernel_size = 3

# 6 3x3 kernels


stride = 1
padding = 0

height, width, channels = 6, 6, 3  # Small 4x4 image with 3 color channels
mock_image = np.random.randint(0, 256, (channels, height, width), dtype=np.uint8)

cnn_model = cnn(input_size, num_kernels, kernel_size)
output = cnn_model.convolve(mock_image, stride, padding)
```
